```python
#histogram for chromosome
import matplotlib.pyplot as plt
import seaborn as sns

df_sorted = df.sort_values('chromosome')

# Assuming 'df' is your DataFrame with columns 'rs_number' and 'chromosome'
plt.figure(figsize=(10, 5))
sns.histplot(data=df, x='chromosome', bins=range(1, 26), kde=False, color='skyblue')
plt.title('Distribution of SNPs Across Chromosomes')
plt.xlabel('Chromosome')
plt.ylabel('Count of SNPs')
plt.show()
```

```python
#histogram for reference_allele
df_sorted = df.sort_values('reference_allele')
plt.figure(figsize=(10, 5))
sns.countplot(data=df, x='reference_allele', color='skyblue', order=df['reference_allele'].value_counts().index)
plt.title('Distribution of Reference Alleles')
plt.xlabel('Reference Allele')
plt.ylabel('Count of SNPs')
plt.show()
```

```python
#beta
df_sorted = df.sort_values('beta')
plt.figure(figsize=(10, 6))
plt.hist(df['beta'], bins=20, color='skyblue', edgecolor='black')
plt.xlabel('Beta Values')
plt.ylabel('Frequency')
plt.title('Distribution of Beta Values')
plt.show()
```

```python
# Group by 'reference_allele' and calculate the mean of 'beta' for each group
grouped_data = df.groupby('reference_allele')['beta'].mean()

```

```python
# Plotting a bar graph
grouped_data.plot(kind='bar', color='skyblue')
plt.xlabel('Reference Allele')
plt.ylabel('Mean Beta Value')
plt.title('Reference Allele vs Mean Beta Value')
plt.show()
```
